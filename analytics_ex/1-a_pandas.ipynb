## DataFrame 실습
1. 딕셔너리를 만들고,
2. 데이터프레임으로 만든 후,
3. nan을 평균값으로 채운다.

```python
import pandas as pd
import numpy as np
# 각 유저별 별점을 주는것이므로, 1 decimal 로 셋팅.
# pd.set_option('precision', 1)

# 책 제목과 작가, 그리고 유저별 별점 데이터가 있다.
books = pd.Series(data = ['Great Expectations', 'Of Mice and Men', 'Romeo and Juliet', 'The Time Machine', 'Alice in Wonderland' ])
authors = pd.Series(data = ['Charles Dickens', 'John Steinbeck', 'William Shakespeare', ' H. G. Wells', 'Lewis Carroll' ])

user_1 = pd.Series(data = [3.2, np.nan ,2.5])
user_2 = pd.Series(data = [5., 1.3, 4.0, 3.8])
user_3 = pd.Series(data = [2.0, 2.3, np.nan, 4])
user_4 = pd.Series(data = [4, 3.5, 4, 5, 4.2])

#  np.nan values 는 해당 유저가 해당 책에는 아직 별점 주지 않은것이다.
# labels: 'Author', 'Book Title', 'User 1', 'User 2', 'User 3', 'User 4'. 
# 아래 그림처럼 나오도록 만든다.
```

```python
# # 1. 딕셔너리를 만들고,     
my_data = {'Book Title' : books,
     'Author' : authors,
     'user_1' : user_1,
     'user_2' : user_2,
     'user_3' : user_3,
     'user_4' : user_4}
```

```python
# 2. 데이터 프레임 만들기
df = pd.DataFrame(my_data)
df
```

```python
# 3. nan을  평균값으로 채운다.
df.fillna(df.mean(numeric_only=True))
```

```python
# 시급이 30보다 이상이면 'A'
# 시급이 30보다 작으면 'B'
# 이렇게 새로운 컬럼 group 컬럼을 만들어서 데이터를 셋팅하세요.
```

```python
df.head()
```

```python
#  구분 기준은 salary >= 30 -> A, 아니면 B
def get_group(Salary):
    if Salary >= 30:
        return 'A'
    elif Salary < 30:
        return 'B'
```

```python
get_group(29)
```

```python
df['Salary [$/h]'].apply(get_group)
```

```python
# group 컬럼을 만들어서 df에 적용하기
df['group'] = df['Salary [$/h]'].apply(get_group)
df.head()
```
