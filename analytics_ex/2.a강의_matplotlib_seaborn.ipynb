# Pie Charts

```python
data= df['generation_id'].value_counts()
```

```python
data
```

```python
plt.pie (data, # generation_id
         labels=data.index, 
         startangle=90, # 시작 위치 각도
         wedgeprops= {'width': 0.7} # 도넛 크기
        )

plt.title('게임 세개별 갯수')
plt.legend(loc = 'upper right')
plt.show()
```

```python
# 한글 표현하기

from matplotlib import font_manager, rc
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우 기본 한글 폰트
plt.rcParams['axes.unicode_minus'] = False      # 마이너스 깨짐 방지

# pie 차트 그리기
plt.pie (data, # generation_id
         labels=data.index, 
         startangle=90, # 시작 위치 각도
         wedgeprops= {'width': 0.7} # 도넛 크기
        )

plt.title('게임 세개별 갯수')
plt.legend(loc = 'upper right')
plt.show()
```

# Bivariate (여러개의 변수) Visualization 방법

```python
data_path = 'datas/fuel_econ.csv'
df = pd.read_csv(data_path)
df
```

```python
df.columns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# 두 컬럼간의 관계를 알고 싶을때.
# 관계 => 상관관계분석   : 피어슨 상관계수 =>  -1 ~ 1 사이의 숫자
# 0 근처는 두 컬럼간에 아무 관계 없다.
# 1 로 갈수록 비례관계다 
# -1 로 갈수록 반비례 관계다.
```

```python
# 상관분석 
df.corr(numeric_only=True)
```

```python
# 배기량과 연비의 관계 => 강한 반비례 관계 
df[['displ','comb']].corr()
```

```python
df.head()
```

```python
plt.scatter(data= df, x= 'displ', y='comb')
plt.title('배기량과 연비의 관계')
plt.xlabel('displ')
plt.ylabel('comb')
plt.show()
```

```python
# 이 차트를 보고, 데이터는 3929개의 데이터 점을 보고서, fitting 되는 선(regression 회귀 )을 상상하면, 오른쪽 아래로 내려가는 
# 선을 그릴수 있다. 따라서 반비례 관계다 라고 알수 있다. 
```

```python
df.head()
```

```python
# fitting 되는 선을 찾는것 => regression
```

```python
sb.regplot(data =df, x='displ', y='comb')
plt.show()
```

```python
# 상관계수도 같이 보여준다. 
df[['displ', 'comb']].corr()
```

```python
sb.pairplot(data= df, vars=['displ', 'comb', 'co2'])
plt.show()
```

# 문제 2. 엔진 크기와 이산화탄소 배출량의 관계를 분석하세요.
- (히트맵으로) displ 이 엔진사이즈이고, co2 가 이산화탄소 배출량입니다. 트랜드를 분석하세요

```python
#  엔진 크기와 이산화탄소 배출량의 관계 분석하기
# (히트맵으로)   displ 이 엔진사이즈이고, co2 가 이산화탄소 배출량입니다.  트랜드 분석하기
```

```python
df.columns
```

```python
city_highway_corr= df[['displ','co2']].corr()
city_highway_corr
```

```python
sb.color_palette()
```

```python
sb.heatmap(city_highway_corr,
          vmax=1.00,
          vmin=0.86,
          cbar = True,
          linewidths=0.01,
          annot=True,
           cmap='Reds'
)
plt.title('엔진크기와 이산화탄소 배출량의 관계')
plt.xlabel('displ')
plt.ylabel('co2')
plt.show()
```
